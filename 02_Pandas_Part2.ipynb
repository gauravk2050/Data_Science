```python
from io import StringIO, BytesIO
```

```python
# StringIO is going to create string memory buffer
```

```python
# creating csv data
data = ('col1,col2,col3\n'
            'x,y,1\n'
            'a,b,2\n'
            'c,d,3')
```

```python
type(data
    )
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from specific columns
df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
df
```

```python
# saving this data frame as csv
df.to_csv('Test.csv')
```

```python
## Specifying columns data types

data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11,12')
```

```python
print(data)
```

```python
# specify column data type
df=pd.read_csv(StringIO(data),dtype=object)
#making data type as an object i.e all values are as string
```

```python
df
```

```python
## Index columns and training delimiters
```

```python
data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data))
```

```python
# giving first column as an index
pd.read_csv(StringIO(data),index_col=0)
```

```python
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data))
```

```python
# above it is using data values as an index so fot avoding this use
pd.read_csv(StringIO(data),index_col=False)
```

```python
## Combining usecols and index_col
data = ('a,b,c\n'
        '4,apple,bat,\n'
        '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data), usecols=['b', 'c'],index_col=False)
```

```python

## Quoting and Escape Characters¶. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```
